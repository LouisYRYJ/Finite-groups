```python
import torch as t
import numpy as np
from matplotlib import pyplot as plt
```

```python
avar, loss, norm = t.load('../data/avar_loss_norm.pt')
```

```python
plt.figure(figsize=(8, 8))
import matplotlib; matplotlib.rc('font', size=24)
plt.scatter(avar, loss, s=70)
plt.xlabel('$a$ variance')
plt.xscale('log')
plt.ylabel('loss')
plt.savefig('../figs/avar_loss.png')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
import matplotlib; matplotlib.rc('font', size=24)
plt.scatter(avar, norm, s=70)
plt.xlabel('$a$ variance')
plt.xscale('log')
plt.ylabel('weight norm')
plt.savefig('../figs/avar_weight.png')
plt.show()
```
